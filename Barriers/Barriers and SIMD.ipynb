```python
import pandas as pd
```

```python
YAF = pd.read_csv('YAF2.csv', encoding='latin-1')
```

```python
YAFNE = YAF[(YAF['Record Type'] == 'New Engagement')]
```

```python
YAFNE
```

```python
YAFNE.Barriers
```

```python
# Remove white spaces from the "Barriers" column
YAFNE['Barriers'] = YAFNE['Barriers'].str.replace(' ', '')

# Create a set to store unique barrier values
unique_barriers = set()

# Iterate over each row in the "Barriers" column
for row in YAFNE['Barriers'].dropna():
    # Split the values in the current row by semicolon
    row_values = row.split(';')
    
    # Add each unique value to the set
    unique_barriers.update(row_values)

# Create new columns based on the unique barrier values
for barrier in unique_barriers:
    YAFNE[barrier] = YAFNE['Barriers'].apply(lambda x: 1 if isinstance(x, str) and barrier in x.split(';') else 0)


# Display the updated DataFrame
YAFNE.head()
```

```python
# Create 'barriers_all' column
YAFNE['barriers_all'] = 0

# List of columns to iterate through
columns_to_count = [
    'MentalHealthChallenges', 'Refugeeorasylumseeker', 'Homeless',
    'LookedafterYP/CareExperienced', 'Criminalconviction', 'Unemployment',
    'Physicalhealthissues/disability', 'Transportationissues', 'FamilyIssues',
    'Carer', 'AddictionIssues', 'Literacy/Numeracyissues', 'Learningchallenges'
]

# Iterate through specified columns and count occurrences of '1'
for column in columns_to_count:
    YAFNE['barriers_all'] += YAFNE[column]

YAFNE.head(10)
```

```python
# Count the occurrences of values in the 'QUINTILE' column
quintile_counts = YAFNE['QUINTILE'].value_counts()

# Group by 'QUINTILE' and count the occurrences of 'barriers_all' in each group
barriers_counts_by_quintile = YAFNE.groupby('QUINTILE')['barriers_all'].value_counts()

# Print the results
print("Counts of values in QUINTILE:")
print(quintile_counts)

print("\nCounts of barriers_all in each QUINTILE:")
print(barriers_counts_by_quintile)
```
